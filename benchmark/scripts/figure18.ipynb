```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Ignore harmless warnings
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
cat_train_cpu = [16.45, 18.28,146.43,173.31,351.99]
cat_train_gpu = [166.62,199.18,302.62,35.56,43.13]
```

```python
labels = ['5K', '10K', '50K','100K','0.5M']
x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars
plt.rcParams['hatch.linewidth'] = 2.5

fig, axs = plt.subplots(figsize = (10, 4))
ax0 = axs
rects0 = ax0.bar(x - width/2, cat_train_cpu, width, label='CPU',linewidth=2.5, color='none', edgecolor=['#1f77b4','#1f77b4'], hatch='')
ax0.bar_label(rects0, padding=3,fontsize=15)
rects1 = ax0.bar(x + width/2, cat_train_gpu, width, label='GPU',linewidth=2.5, color='none', edgecolor=['#ff7f0e','#ff7f0e'], hatch='/')
ax0.bar_label(rects1, padding=3,fontsize=15)
ax0.set_xticklabels(['','5K', '10K', '50K','100K','0.5M'])

ax0.ticklabel_format(style='sci',axis='y')
# percentage = [int((a - b) / a * 100) for a,b in zip(cat_train_cpu,cat_train_gpu)]
# for x,y,z in zip(x,cat_train_cpu,cat_train_gpu):
#     ax0.text(x - width/2,y + 1,str(y),ha='center',fontsize=10)
#     ax0.text(x + width/2,z + 9,str(z),ha='center',fontsize=10)
ax0.legend(fontsize=15,loc='upper left')
ax0.grid(alpha=0.5)
ax0.tick_params(axis='both', which='major', labelsize=15)
# ax0.set_yscale("log")
ax0.set_ylim(top=400)
ax0.set_ylabel('Training Time (s)', fontsize=15)
# fig.savefig('./cat_gpu_time.pdf')
```
